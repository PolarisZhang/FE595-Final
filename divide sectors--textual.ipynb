```python
import pandas as pd
import csv
```

```python
## change input & output file name
df=pd.read_csv('RFactors_2019std.csv')   ## change input file name

c1=[]  ## ticker
c2=[]
c3=[]
c4=[]
c5=[]
c6=[]
c7=[]
c8=[]
c9=[]
c10=[]

m1=[]   ## pc1
m2=[]
m3=[]
m4=[]
m5=[]
m6=[]
m7=[]
m8=[]
m9=[]
m10=[]

b1=[]   ## pc2
b2=[]
b3=[]
b4=[]
b5=[]
b6=[]
b7=[]
b8=[]
b9=[]
b10=[]

r1=[]   ## std.return
r2=[]
r3=[]
r4=[]
r5=[]
r6=[]
r7=[]
r8=[]
r9=[]
r10=[]


c=[]
m=[]
b=[]
r=[]

## divide company condition on SIC range
for i in range(0,len(df)):
    if df.SIC[i]>=100 and df.SIC[i]<=900:     # argriculture, forestry, fishing
        c1.append(df.ticker[i])
        m1.append(df.pc1[i])
        b1.append(df.pc2[i])
        r1.append(df['std'][i])
    elif df.SIC[i]>=1000 and df.SIC[i]<=1400:  # mining
        c2.append(df.ticker[i])
        m2.append(df.pc1[i])
        b2.append(df.pc2[i])
        r2.append(df['std'][i])
    elif df.SIC[i]>=1500 and df.SIC[i]<=1700:  # construction
        c3.append(df.ticker[i])
        m3.append(df.pc1[i])
        b3.append(df.pc2[i])
        r3.append(df['std'][i])
    elif df.SIC[i]>=2000 and df.SIC[i]<=3900:  # manufacturing
        c4.append(df.ticker[i])
        m4.append(df.pc1[i])
        b4.append(df.pc2[i])
        r4.append(df['std'][i]) 
    elif df.SIC[i]>=4000 and df.SIC[i]<=4900:  # transportation, communications,electric, gas, sanitary services
        c5.append(df.ticker[i])
        m5.append(df.pc1[i])
        b5.append(df.pc2[i])
        r5.append(df['std'][i])
    elif df.SIC[i]>=5000 and df.SIC[i]<=5100:  # wholesale trade
        c6.append(df.ticker[i])
        m6.append(df.pc1[i])
        b6.append(df.pc2[i])
        r6.append(df['std'][i])
    elif df.SIC[i]>=5200 and df.SIC[i]<=5900:  # retail trade
        c7.append(df.ticker[i])
        m7.append(df.pc1[i])
        b7.append(df.pc2[i])
        r7.append(df['std'][i])
    elif df.SIC[i]>=6000 and df.SIC[i]<=6700:  # finance insurance, real estate
        c8.append(df.ticker[i])
        m8.append(df.pc1[i])
        b8.append(df.pc2[i])
        r8.append(df['std'][i])
    elif df.SIC[i]>=7000 and df.SIC[i]<=8900:  # services
        c9.append(df.ticker[i])
        m9.append(df.pc1[i])
        b9.append(df.pc2[i])
        r9.append(df['std'][i])
    elif df.SIC[i]>=9000 and df.SIC[i]<=9900:  # public administration
        c10.append(df.ticker[i])
        m10.append(df.pc1[i])
        b10.append(df.pc2[i])
        r10.append(df['std'][i])
    
c.append(c1)  ## record ticker
c.append(c2)
c.append(c3)
c.append(c4)
c.append(c5)
c.append(c6)
c.append(c7)
c.append(c8)
c.append(c9)
c.append(c10)

m.append(m1)  ## record pc1
m.append(m2)
m.append(m3)
m.append(m4)
m.append(m5)
m.append(m6)
m.append(m7)
m.append(m8)
m.append(m9)
m.append(m10)

b.append(b1)  ## record pc2
b.append(b2)
b.append(b3)
b.append(b4)
b.append(b5)
b.append(b6)
b.append(b7)
b.append(b8)
b.append(b9)
b.append(b10)

r.append(r1)  ## std. return
r.append(r2)
r.append(r3)
r.append(r4)
r.append(r5)
r.append(r6)
r.append(r7)
r.append(r8)
r.append(r9)
r.append(r10)

# change output file name
for i in range(0,9):
    k=1+i
    with open('T2019_Sec'+str(k)+'_std.csv', 'w', encoding='utf-8', newline='') as f:
        csvwriter = csv.writer(f)
        csvwriter.writerow(['ticker','pc1','pc2','std'])
        for j in range(0,len(c[i])):
            csvwriter.writerow([c[i][j],m[i][j],b[i][j],r[i][j]])

```
