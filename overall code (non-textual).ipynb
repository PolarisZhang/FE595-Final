```python
import pandas as pd
import csv

## change input & output file name
df=pd.read_csv('2019_cal.csv')   ## change input file name

c1=[]
c2=[]
c3=[]
c4=[]
c5=[]
c6=[]
c7=[]
c8=[]
c9=[]
c10=[]

m1=[]
m2=[]
m3=[]
m4=[]
m5=[]
m6=[]
m7=[]
m8=[]
m9=[]
m10=[]

b1=[]
b2=[]
b3=[]
b4=[]
b5=[]
b6=[]
b7=[]
b8=[]
b9=[]
b10=[]


c=[]
m=[]
b=[]
for i in range(0,len(df)):
    if df.SIC[i]>=100 and df.SIC[i]<=900:     # argriculture, forestry, fishing
        c1.append(df.ticker[i])
        m1.append(df.marketCap[i])
        b1.append(df.BTM[i])
    elif df.SIC[i]>=1000 and df.SIC[i]<=1400:  # mining
        c2.append(df.ticker[i])
        m2.append(df.marketCap[i])
        b2.append(df.BTM[i])
    elif df.SIC[i]>=1500 and df.SIC[i]<=1700:  # construction
        c3.append(df.ticker[i])
        m3.append(df.marketCap[i])
        b3.append(df.BTM[i])
    elif df.SIC[i]>=2000 and df.SIC[i]<=3900:  # manufacturing
        c4.append(df.ticker[i])
        m4.append(df.marketCap[i])
        b4.append(df.BTM[i])
    elif df.SIC[i]>=4000 and df.SIC[i]<=4900:  # transportation, communications,electric, gas, sanitary services
        c5.append(df.ticker[i])
        m5.append(df.marketCap[i])
        b5.append(df.BTM[i])
    elif df.SIC[i]>=5000 and df.SIC[i]<=5100:  # wholesale trade
        c6.append(df.ticker[i])
        m6.append(df.marketCap[i])
        b6.append(df.BTM[i])
    elif df.SIC[i]>=5200 and df.SIC[i]<=5900:  # retail trade
        c7.append(df.ticker[i])
        m7.append(df.marketCap[i])
        b7.append(df.BTM[i])
    elif df.SIC[i]>=6000 and df.SIC[i]<=6700:  # finance insurance, real estate
        c8.append(df.ticker[i])
        m8.append(df.marketCap[i])
        b8.append(df.BTM[i])
    elif df.SIC[i]>=7000 and df.SIC[i]<=8900:  # services
        c9.append(df.ticker[i])
        m9.append(df.marketCap[i])
        b9.append(df.BTM[i])
    elif df.SIC[i]>=9000 and df.SIC[i]<=9900:  # public administration
        c10.append(df.ticker[i])
        m10.append(df.marketCap[i])
        b10.append(df.BTM[i])
    
c.append(c1)
c.append(c2)
c.append(c3)
c.append(c4)
c.append(c5)
c.append(c6)
c.append(c7)
c.append(c8)
c.append(c9)
c.append(c10)

m.append(m1)
m.append(m2)
m.append(m3)
m.append(m4)
m.append(m5)
m.append(m6)
m.append(m7)
m.append(m8)
m.append(m9)
m.append(m10)

b.append(b1)
b.append(b2)
b.append(b3)
b.append(b4)
b.append(b5)
b.append(b6)
b.append(b7)
b.append(b8)
b.append(b9)
b.append(b10)

# change output file name
for i in range(0,9):
    with open('2020_'+str(i)+'.csv', 'w', encoding='utf-8', newline='') as f:
        csvwriter = csv.writer(f)
        csvwriter.writerow(['ticker','marketCap','BTM'])
        for j in range(0,len(c[i])):
            csvwriter.writerow([c[i][j],m[i][j],b[i][j]])

```

## divide each sector by market cap & B/M

```python
import pandas as pd
import csv

from pandas_datareader import data as pdr
import yfinance as yf
from datetime import datetime

import statistics
import numpy as np
from operator import itemgetter
```

```python
## change input file name
df=pd.read_csv('2018_Sec1.csv')   ## year:15-19   sector:1-9

# seperate to small & big by mid market cap
med=[int(i) for i in df.marketCap]
med=statistics.median(med)

b1=[]
b2=[]
m1=[]
m2=[]
t1=[]
t2=[]
for i in range(0,len(df)):
    if df.marketCap[i]<=med:
        b1.append(df.BTM[i])
        m1.append(df.marketCap[i])
        t1.append(df.ticker[i])
    else:
        b2.append(df.BTM[i])
        m2.append(df.marketCap[i])
        t2.append(df.ticker[i])
        
# in small portfolio, seperate into 3 by btm (b1)
a=[]
b1=[int(i) for i in b1]
#t1=[int(i) for i in t1]
m1=[int(i) for i in m1]
for i in range(0,len(b1)):
    temp=[]
    temp.append(t1[i])
    temp.append(m1[i])
    temp.append(b1[i])
    a.append(temp)
a = sorted(a, key=itemgetter(2))

count1=int(len(a)/3)
count2=int((len(a)/3)*2)


# in big portfolio, seperate into 3 by btm (b1)
s=[]
b2=[int(i) for i in b2]
#t1=[int(i) for i in t1]
m2=[int(i) for i in m2]
for i in range(0,len(b2)):
    temp=[]
    temp.append(t2[i])
    temp.append(m2[i])
    temp.append(b2[i])
    s.append(temp)
s = sorted(s, key=itemgetter(2))

count3=int(len(s)/3)
count4=int((len(s)/3)*2)
```

```python
# change output file name

year=18 ## year (15-19)
k=1     ## sector (1-9)


with open('20'+str(year)+'_Sec'+str(k)+'_1.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(0,count1):
        csvwriter.writerow([a[i][0],a[i][1],a[i][2]])
        
with open('20'+str(year)+'_Sec'+str(k)+'_2.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(count1,count2):
        csvwriter.writerow([a[i][0],a[i][1],a[i][2]])
            
with open('20'+str(year)+'_Sec'+str(k)+'_3.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(count2,len(a)):
        csvwriter.writerow([a[i][0],a[i][1],a[i][2]])
            
with open('20'+str(year)+'_Sec'+str(k)+'_4.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(0,count3):
        csvwriter.writerow([s[i][0],s[i][1],s[i][2]])
                
with open('20'+str(year)+'_Sec'+str(k)+'_5.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(count3,count4):
        csvwriter.writerow([s[i][0],s[i][1],s[i][2]])

with open('20'+str(year)+'_Sec'+str(k)+'_6.csv', 'w', encoding='utf-8', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(['ticker','marketCap','BTM'])
    for i in range(count4,len(s)):
        csvwriter.writerow([s[i][0],s[i][1],s[i][2]])
```
