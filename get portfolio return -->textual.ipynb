```python
# for single item
yf.pdr_override() 


df=pd.read_csv('T2015_Sec2_std1_pc1#1.csv')   ## input dataset
tickers=list(df.ticker)     

# get monthly adj. close price from yahoo finance (for each stock) using ticker
adj_close = pdr.get_data_yahoo(tickers,start = "2014-12-31",end = "2015-12-31")['Adj Close'].resample('M').ffill()

# monthly return for each stock
returns=adj_close.pct_change()
    
## find each stock's annual return first, then find portfolio return
#s_ann_return=returns.mean()    # stock annual return
#s_ann_return.mean()            # portfolio annual return from all stock annual return

## portfolio ann return
p_ann_return=returns.stack().mean()
#p_ann_return=returns.mean()        # for portfolio with only 1 column
p_ann_return   
```
